```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data['robot']=data['whoAmI'].eq('robot').map({True:1,False:0})
data['human']=data['whoAmI'].eq('human').map({True:1,False:0})

del data['whoAmI']
data
```

```python

fig, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(data, annot=True, cmap='GnBu', linewidths=.3, linecolor='black', cbar=False);
```
